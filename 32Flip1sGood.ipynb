```python
'''
https://www.interviewbit.com/problems/flip/
You are given a binary string A(i.e. with characters 0 and 1) consisting of characters A1, A2, ..., AN. In a single operation, you can choose two indices L and R such that 1 ≤ L ≤ R ≤ N and flip the characters AL, AL+1, ..., AR. By flipping, we mean change character 0 to 1 and vice-versa.

Your aim is to perform ATMOST one operation such that in final string number of 1s is maximised.

If you don't want to perform the operation, return an empty array. Else, return an array consisting of two elements denoting L and R. If there are multiple solutions, return the lexicographically smallest pair of L and R.

NOTE: Pair (a, b) is lexicographically smaller than pair (c, d) if a < c or, if a == c and b < d.
'''

#0111000100010


def flip( A):
    
    arr = [int(x) for x in A]           # construct an array from string
    n, initSum = len(arr), sum(arr)
    
    if n == initSum:        # there are no 0s present
        return []
        
    left,right = -1, -1
    for i in range(n):
        if arr[i] == 0:
            left,right = i, i           # find first 0 and set left, right to it
            break
    
    currSum = initSum
    maxPair = []
    maxSum = -1
    
    while right < n:        # this is a modified kadane
        print(left,right,currSum)
        if arr[right] == 0:     # we will add 1 because we will flip the zero
            currSum += 1
            if currSum > maxSum:        # check if the current flipped is max
                maxSum = currSum
                maxPair = [left+1,right+1]
            right += 1
        elif arr[right] == 1:
            currSum -= 1    # subtract 1 from cursum as 1 -> 0
            if currSum > initSum:
                right += 1  # if currSum is still greater than sum of array, we can continue without tension
            else:
                while right < n:    # now we have to move left, right to the first zero after these 1s
                    if arr[right] == 1:
                        right += 1
                    else:
                        break
                left = right
                currSum = initSum
    
    return maxPair
    
print(flip("0111000100010"))    
```
